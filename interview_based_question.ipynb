```python
#python

#1

def find_highest_frequency_word_length(string):
    word_frequency = {}
    words = string.split()

    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1

    highest_frequency = 0
    highest_frequency_word_length = 0

    for word, frequency in word_frequency.items():
        if frequency > highest_frequency:
            highest_frequency = frequency
            highest_frequency_word_length = len(word)
        elif frequency == highest_frequency and len(word) > highest_frequency_word_length:
            highest_frequency_word_length = len(word)

    return highest_frequency_word_length


string = "write write write all the number from from from 1 to 100"
result = find_highest_frequency_word_length(string)
print(result)

```

```python
#2


from collections import Counter

def is_valid_string(s):
    char_count = Counter(s)
    char_freq = Counter(char_count.values())

    if len(char_freq) == 1:
        return "YES"

    if len(char_freq) == 2:
        min_freq, max_freq = sorted(char_freq.keys())

        if char_freq[min_freq] == 1 and min_freq == 1:
            return "YES"

        if char_freq[max_freq] == 1 and max_freq - min_freq == 1:
            return "YES"

    return "NO"


s1 = "abc"
result1 = is_valid_string(s1)
print(result1) 

s2 = "abcc"
result2 = is_valid_string(s2)
print(result2) 
```

```python
#3

import requests
import pandas as pd

def download_data(url):
    response = requests.get(url)
    data = response.json()
    return data

def convert_to_dataframe(data):
    attributes = ['id', 'num', 'name', 'img', 'type', 'height', 'weight', 'candy', 'candy_count', 'egg',
                  'spawn_chance', 'avg_spawns', 'spawn_time', 'multipliers', 'weaknesses', 'next_evolution', 'prev_evolution']
    
    extracted_data = {attr: [] for attr in attributes}
    
    for pokemon in data['pokemon']:
        for attr in attributes:
            extracted_data[attr].append(pokemon.get(attr, ''))
    
    
    df = pd.DataFrame(extracted_data)
    
    return df

def export_to_excel(df, filename):
    df.to_excel(filename, index=False)

url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
pokemon_data = download_data(url)

df_pokemon = convert_to_dataframe(pokemon_data)

filename = 'pokemon_data.xlsx'
export_to_excel(df_pokemon, filename)

print("Data has been downloaded, converted, and saved to", filename)
```

```python
#5

import requests

def download_data(url):
    response = requests.get(url) 
    response.raise_for_status()  
    data = response.json()  
    return data

def extract_episodes_data(data):
    episodes_data = []
    episodes = data['_embedded']['episodes'] 
    for episode in episodes:
        episode_data = {
            'id': episode['id'],
            'url': episode['url'],
            'name': episode['name'],
            'season': episode['season'],
            'number': episode['number'],
            'type': episode['type'],
            'airdate': episode['airdate'],
            'airtime': episode['airtime'],
            'runtime': episode['runtime'],
            'average_rating': episode['rating']['average'],
            'summary': episode['summary'].strip('<p>').strip('</p>'),
            'medium_image': episode['image']['medium'],
            'original_image': episode['image']['original'],
            'summary': episode['summary'].strip('<p>').strip('</p>'),
            'links': {
                'self': episode['_links']['self']['href'],
                'show': episode['_links']['show']['href']
            }
        }
        episodes_data.append(episode_data)
    return episodes_data

url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

data = download_data(url)

episodes_data = extract_episodes_data(data)

for episode_data in episodes_data:
    print("Episode:")
    for key, value in episode_data.items():
        if isinstance(value, dict):
            print(f"{key}:")
            for sub_key, sub_value in value.items():
                print(f"  {sub_key}: {sub_value}")
        else:
            print(f"{key}: {value}")
    print()
```

```python
#6


import requests
import matplotlib.pyplot as plt

def download_data(url):
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()
    return data

def get_pokemon_with_low_spawn_rate(pokemon_data):
    return [pokemon for pokemon in pokemon_data if pokemon['spawn_chance'] < 0.05]

def get_pokemon_with_few_weaknesses(pokemon_data):
    return [pokemon for pokemon in pokemon_data if len(pokemon.get('weaknesses', [])) < 4]

def get_pokemon_with_no_multipliers(pokemon_data):
    return [pokemon for pokemon in pokemon_data if 'multipliers' not in pokemon]

def get_pokemon_with_few_evolutions(pokemon_data):
    return [pokemon for pokemon in pokemon_data if len(pokemon.get('evolution', [])) <= 2]

def get_pokemon_with_low_spawn_time(pokemon_data):
    return [pokemon for pokemon in pokemon_data if pokemon['spawn_time'] != 'N/A' and 
            int(pokemon['spawn_time'].split(':')[0]) * 60 +
            int(pokemon['spawn_time'].split(':')[1]) < 300]

def get_pokemon_with_multiple_capabilities(pokemon_data):
    return [pokemon for pokemon in pokemon_data if len(pokemon.get('capabilities', [])) > 2]

url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"

data = download_data(url)
pokemon_data = data['pokemon']

pokemon_low_spawn_rate = get_pokemon_with_low_spawn_rate(pokemon_data)
print("Pokemon with low spawn rate:")
for pokemon in pokemon_low_spawn_rate:
    print(pokemon['name'])
print()

pokemon_few_weaknesses = get_pokemon_with_few_weaknesses(pokemon_data)
print("Pokemon with few weaknesses:")
for pokemon in pokemon_few_weaknesses:
    print(pokemon['name'])
print()

pokemon_no_multipliers = get_pokemon_with_no_multipliers(pokemon_data)
print("Pokemon with no multipliers:")
for pokemon in pokemon_no_multipliers:
    print(pokemon['name'])
print()

pokemon_few_evolutions = get_pokemon_with_few_evolutions(pokemon_data)
print("Pokemon with few evolutions:")
for pokemon in pokemon_few_evolutions:
    print(pokemon['name'])
print()

pokemon_low_spawn_time = get_pokemon_with_low_spawn_time(pokemon_data)
print("Pokemon with low spawn time:")
for pokemon in pokemon_low_spawn_time:
    print(pokemon['name'])
print()

```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
#10

def count_pos_tags(text):
   words = word_tokenize(text)
    
   pos_tags = pos_tag(words)
    
   verb_count = 0
   noun_count = 0
   pronoun_count = 0
   adjective_count = 0
    
   for word, tag in pos_tags:
        if tag.startswith('V'):
            verb_count += 1
        elif tag.startswith('N'):
            noun_count += 1
        elif tag.startswith('PRP'):
            pronoun_count += 1
        elif tag.startswith('JJ'):
            adjective_count += 1
    
   pos_counts = {
        'NOUN': noun_count,
        'PRONOUN': pronoun_count,
        'VERB': verb_count,
        'ADJECTIVE': adjective_count
    }
    
   return pos_counts

phrase = "The cat jumped over the lazy dog"
result1 = count_pos_tags(phrase)
print("Count of POS tags in phrase:")
print(result1)

paragraph = "She walked to the park and saw a beautiful flower. It smelled lovely."
result2 = count_pos_tags(paragraph)
print("Count of POS tags in paragraph:")
print(result2)

text = "The quick brown fox jumps over the lazy dog. The dog barks loudly."
result3 = count_pos_tags(text)
print("Count of POS tags in text:")
print(result3)
```
